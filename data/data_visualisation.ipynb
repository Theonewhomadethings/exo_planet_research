```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import csv
```

```python
file = open('structured_data.csv')
type(file)
csvreader = csv.reader(file)
header = []
header = next(csvreader)
header
```

```python
rows = []
column_1 = []
column_2 = []
column_3 = []


for row in csvreader:
    rows.append(row)
    column_1.append(float(row[0]))
    column_2.append(float(row[1]))
    column_3.append(float(row[2]))
    
```

```python
print(column_1)
```

```python
x = column_1 #JD double days time
y = column_2 # RV double  velocity

plt.plot(x, y)
plt.xlabel("JD double days time")
plt.ylabel("RV double velocity")
plt.show()
```

```python
scaler = MinMaxScaler()
column_1_normalized = scaler.fit_transform([[value] for value in column_1])

x_norm = column_1_normalized.flatten()

plt.plot(x_norm, y)
plt.xlabel("Normalised JD double days time")
plt.ylabel("RV double velocity")
plt.show()
```

```python
x = column_1 #JD double days time
y = column_2 # RV double  velocity

plt.plot(x, y)
plt.xlabel("JD double days time")
plt.ylabel("RV double velocity")
plt.xscale('log')
plt.show()
```

```python
data = []
with open('structured_data.csv', 'r') as file:
    csvreader = csv.reader(file)
    next(csvreader) #skip the header row
    for row in csvreader:
        data.append(float(row[0]))
        
hist, bins = np.histogram(data, bins = 10)

plt.hist(data, bins = 10)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()
```
